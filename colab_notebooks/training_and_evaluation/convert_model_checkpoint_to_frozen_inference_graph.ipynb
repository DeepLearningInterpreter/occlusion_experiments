##Introduction
This notebook can be used after training a model. The training checkpoint that is specified will be converted to a "frozen_inference_graph.pb" file, which is useful for inference. This file will enable you to conduct occlusion experiments on the newly trained model using the "Creating_heat_maps.ipynb" notebook.

In short, this notebook produces a "frozen_inference_graph.pb" file from a model checkpoint. The resulting file must be saved so it can be used later.

##Choose settings
In the next codeblock the model can be specified of which the checkpoint needs to be converted. Additionally, it can be specified where the checkpoint is located and where the resulting files should be outputted. It is important to specify the number of the checkpoint you want to convert to a frozen inference graph with the *ckpt_number* variable. 

```
which_model = "SSD_ext_lrCyc"
ckpt_number = "22700"

PIPELINE_CONFIG_PATH = "/content/occlusion_experiments/main_content/multitude_of_possible_detectors/training/" + which_model + "/pipeline.config" 
TRAINED_CHECKPOINT = "/content/occlusion_experiments/main_content/multitude_of_possible_detectors/training/" + which_model + "/model.ckpt-" + ckpt_number
OUTPUT_DIRECTORY = "/content/occlusion_experiments/main_content/multitude_of_possible_detectors/frozen_mosels_for_detection/" + which_model
```
